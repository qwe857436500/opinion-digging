# 经过一步剪枝处理后，仍然有很多无意义的特征词，使用tf-idf去掉，去掉同类文本和目标文本出现评率都低的特征词

## 使用snownlp库获得文本的tf-idf值

```python
from snownlp import SnowNLP
```

## 读取文件

```python
def readFile(fileName):
    word_line1 = []
    with open(fileName,encoding='UTF-8') as f:
        for line in f.readlines():   
            line = line.strip().replace('\n','')
            word_line1.append(line)
    #print(word_line1)
    return word_line1
```

## 使用snownlp进行分词

```python
def getSegText(word_line1):
    words_list = []
    for w in word_line1:
        #print(w)
        s = SnowNLP(w)
        #print(s.words)
        for w in s.words:
            words_list.append(w)
    #print(words_list)
    return words_list
```

```python
word_line1 = readFile(r'C:/Users/Administrator/Desktop/180509/'+'37641160928.txt')
words_list = getSegText(word_line1)


v = SnowNLP([words_list])
#print(v.tf)
print(type(v.tf))
v.tf[0]['物流']
```

## 读取data5中的所有同类文件

## 查看每个特征词的tf-idf值

```python
import pickle
key_list = []
special_fea_list = []
new_fea_list1 = []
with open('F:/data/new_fea_to_opinion_dict.pkl', 'rb') as f:
    new_dict = pickle.load(f)
for key in new_dict:
    key_list.append(key)
    if key in vv.idf.keys():
        print(key)
        print(vv.idf[key])
        if vv.idf[key] > 0 and vv.idf[key] < 1:
            print('这词很特别有问题 idf大于0',key)
            special_fea_list.append(key)
        elif vv.idf[key] > 1:
            print('这个词真的很有问题11111111111111111111111,idf大于1',key)
        else:
            new_fea_list1.append(key)
    else:
        print('其他文档中没有这个词',key)
print(key_list)
print(special_fea_list)
print(new_fea_list1)
```

## 计算tf/idf的值

```python
score_list = []
fea_list = []
for w in special_fea_list:
    if w in vv.tf[5]:
        a = vv.tf[5][w]
        b = vv.idf[w]
        c = a /b
        print(w,c)
        fea_list.append(w)
        score_list.append(c)
```

## 归一化后值大于0.5的留下

```python
from sklearn import preprocessing
import numpy as np
```

```python
score_list = np.array(score_list).reshape(1,-1)
x_nor = preprocessing.normalize(score_list, norm='l2')
print(x_nor)
i = 0
for w in x_nor[0]:
    if w > 0.5:
        print(w,i)
        new_fea_list1.append(fea_list[i])
    i += 1
print(new_fea_list1)
```
