```python
from gensim.models import word2vec
import pickle
model = word2vec.Word2Vec.load('F:/word2vec/word2vec_wx')

with open('F:/data/new_new_fea_to_opinion_dict.pkl', mode='rb') as f:
    fea_to_opinion_dict = pickle.load(f)

with open('F:/data/new_new_opinion_to_fea_dict.pkl', mode='rb') as f:
    opinion_to_fea_dict = pickle.load(f)
    
with open('F:/data3/comments_clean_wordFreq.pkl', mode='rb') as f:
    words_freq = pickle.load(f)
```

```python
def read_lines(filename):
    with open(filename) as f:
        lines = []
        for line in f:
            line = line.strip().replace('\n', '')
            lines.append(line)
        return lines
```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import numpy as np
from math import log
from math import fabs
from collections import defaultdict
import operator
```

# 聚类

```python
def cluster(dataFile, n_clusters):
    words_list = read_lines(dataFile)
    new_words_list = []
    #print(len(words_list))
    #print(type(model.vocabulary))
    #help(model.vocabulary)
    #help(model.vocabulary.prepare_vocab)
    words = [k for k, v in model.wv.vocab.items()]
    
    vector = list()
    for word in words_list:
        if word in words:
            #print(word)
            vector.append(model[word])
            new_words_list.append(word)
    #print(vector)
    kmeans = KMeans(n_clusters, random_state=0).fit(vector)
    #print('labels', kmeans.labels_)
    labels = kmeans.labels_
#     center = kmeans.cluster_centers_
#     print(center)
    return list(labels), new_words_list
    
    
```

```python
opi_lables, opinion_words_list = cluster('F:/data2/candidateFeaOpinion.txt', 20)
```

```python
from scipy import spatial
def sim(word1, word2):
    vector1 = model[word1]
    vector2 = model[word2]
    return spatial.distance.cosine(vector1, vector2)
```

```python
words = [k for k, v in model.wv.vocab.items()]
fea_strength = [1.0 for x in range(0, len(fea_words_list))]
opinion_stren = [1.0 for x in range(0, len(opinion_words_list))]
print(fea_strength)

flag = True
alpha = 0.6
n = 1
tem_fea_strengh = fea_strength
tem_opinion_stren = opinion_stren

while flag and n <= 150:
    print('iter', n)
    n += 1
    #特征类别
    for labels in range(0, 20):
        cur_class_fea = []
        for id, word in enumerate(fea_words_list):
            if fea_lables[id] == labels:
                cur_class_fea.append(word)
        #计算当前类别特征强度
        for curr_fea in cur_class_fea:
            curr_fea_id = fea_words_list.index(curr_fea)
            
            
            stren_from_fea = 0
            #从其他特征获得的强度
            for other_fea in cur_class_fea:
                if other_fea != curr_fea:
                    other_fea_id = fea_words_list.index(other_fea)
                    stren_from_fea += sim(other_fea, curr_fea) * fea_strength[other_fea_id] 
                    
                    
            stren_from_opinion = 0
            #从其他观点词获得的强度
            for curr_fea_opinion in set(fea_to_opinion_dict[curr_fea]):
                if curr_fea_opinion in words and curr_fea_opinion in opinion_words_list:
                    fea_opinion_cocurrence = fea_to_opinion_dict[curr_fea].count(curr_fea_opinion)
                    fea_freq = len(fea_to_opinion_dict[curr_fea])
                    opi_freq = len(opinion_to_fea_dict[curr_fea_opinion])
                    opinion_id = opinion_words_list.index(curr_fea_opinion)
                    stren_from_opinion += log(fea_opinion_cocurrence * 1.0 / fea_freq * opi_freq ) * opinion_stren[opinion_id]
                    
            tem_fea_strengh[curr_fea_id] = alpha * stren_from_fea + (1-alpha) * stren_from_opinion
    #观点类别        
    for labels in range(0, 20):
        cur_class_opi = []
        for id, word in enumerate(opinion_words_list):
            if opi_lables[id] == labels:
                cur_class_opi.append(word)
        
        for curr_opi in cur_class_opi:
            curr_opi_id = opinion_words_list.index(curr_opi)
            
            
            stren_from_opi = 0
            #从其他观点词获得观点词强度
            for other_opi in cur_class_opi:
                if other_opi != curr_opi:
                    other_opi_id = opinion_words_list.index(other_opi)
                    stren_from_opi += sim(other_opi, curr_opi) * opinion_stren[other_opi_id]
            
            
            stren_from_fea = 0
            #从特征词获取观点词强度
            for curr_opinion_fea in set(opinion_to_fea_dict[curr_opi]):
                if curr_opinion_fea in words and curr_opinion_fea in fea_words_list:
                    opinion_fea_cocurrence = opinion_to_fea_dict[curr_opi].count(curr_opinion_fea)
                    #opinion_freq = words_freq[curr_opi]
                    #fea_freq = words_freq[curr_opinion_fea]
                    fea_freq = len(fea_to_opinion_dict[curr_opinion_fea])
                    opi_frea = len(opinion_to_fea_dict[curr_opi])
                    fea_id = fea_words_list.index(curr_opinion_fea)
                    stren_from_fea += log(opinion_fea_cocurrence * 1.0 / fea_freq * opi_freq) * fea_strength[fea_id]
                    
            tem_opinion_stren[curr_opi_id] = alpha * stren_from_opi +(1 - alpha) * stren_from_fea
    
    #L2正则化
    tem_fea_strengh = np.array(tem_fea_strengh).reshape(1, -1)
    tem_opinion_stren = np.array(tem_opinion_stren).reshape(1, -1)
    tem_fea_strengh = preprocessing.normalize(tem_fea_strengh, norm='l2')[0].tolist()
    tem_opinion_stren = preprocessing.normalize(tem_opinion_stren, norm='l2')[0].tolist()
    
    #判断是否收敛
    error = 1e-4
    fea_error = [True if fabs(stre1-stre2) <= error else False for stre1, stre2 in zip(fea_strength, tem_fea_strengh)]
    opi_error = [True if fabs(stre1-stre2) <= error else False for stre1, stre2 in zip(opinion_stren, tem_opinion_stren)]
    
    if False not in fea_error and False not in opi_error:
        print('收敛')
        print('old fea', fea_strength[0:10])
        print('fea', tem_fea_strengh[0:10])
        print('old opinion', opinion_stren[0:10])
        print('opinion', tem_opinion_stren[0:10])
    else:
        fea_strength = tem_fea_strengh
        opinion_stren = tem_opinion_stren
        fea_strength = np.array(fea_strength).reshape(1, -1)
        opinion_stren = np.array(opinion_stren).reshape(1, -1)
        fea_strength = preprocessing.normalize(fea_strength, norm='l2')[0].tolist()
        opinion_stren = preprocessing.normalize(opinion_stren, norm='l2')[0].tolist()
        
        print('fea_strength', fea_strength[0:5])
        print('opinion_strength', opinion_stren[0:5])

feaStrenDict = defaultdict(float)
opinionStrenDict = defaultdict(float)

for word, strength in zip(fea_words_list, fea_strength):
    feaStrenDict[word] = strength
    
for word, strength in zip(opinion_words_list, opinion_stren):
    opinionStrenDict[word] = strength
    
feaStrenDict_sort = sorted(feaStrenDict.items(), key = operator.itemgetter(1), reverse= True)
opinionStrenDict_sort = sorted(opinionStrenDict.items(), key = operator.itemgetter(1), reverse= True)

with open('F:/data4/feaStren.txt', 'w') as f:
    for word, stren in feaStrenDict_sort:
        f.write(word + ':' + str(stren) + '\n')
        
with open('F:/data4/opiStren.txt', 'w') as f:
    for word, stren in opinionStrenDict_sort:
        f.write(word + ':' + str(stren) + '\n')
        
```
