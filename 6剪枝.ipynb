# 观察得到的词对，里面有很多我们不想要的词对，因此要进行剪枝
## 观察到提取的部分观点词不在情感字典中

```python
from collections import defaultdict
import pickle
with open('F:/data2/opinion_to_fea.pkl', 'rb') as f:
    opinion_to_fea_dict = pickle.load(f)
#print(opinion_to_fea_dict)
print(opinion_to_fea_dict.keys())
with open('F:/data2/fea_to_opinion.pkl', 'rb') as f:
    fea_to_opinion_dict = pickle.load(f)
```

## 载入词性字典

```python
with open('F:/data3/comments_clean_posDict.pkl', 'rb') as f:
    pos_dict = pickle.load(f)
print(pos_dict['便宜'])
```

```python
#所有观点词
list1 = []
#在词性字典中出现的形容词
list2 = []
#既存在观点词中又存在于词性字典中打的词
list3 = []
for key in opinion_to_fea_dict:
    list1.append(key)
#print(list1)

for key in pos_dict:
    list2.append(key)
#print(list2)

for w1 in list1:
    for w2 in list2:
        if w2 == w1:
            list3.append(w2)
#情感字典中的词            
lines = [] 
with open(r'C:/Users/Administrator/Desktop/motion/Sentiment_dict/emotion_dict/pos_all_dict.txt', encoding='UTF-8') as f: 
    for line in f.readlines():
        line = line.strip()
        lines.append(line)
#print(lines)
with open(r'C:/Users/Administrator/Desktop/motion/Sentiment_dict/emotion_dict/neg_all_dict.txt', encoding='UTF-8') as f: 
    for line in f.readlines():
        line = line.strip()
        lines.append(line)
        
#不是情感词的动词
list4 = []
for w in list3:
    if pos_dict[w] == 'v' and w not in lines:
        list4.append(w) #list4是要去除的不是感情词的动词
        
```

## list4是要去除的不是情感词的动词

```python
list4#去除的opinion
```

```python
set(list3) - set(list4)#剩下的opinion
```

## 剪枝观点词对应的特征词

```python
del_list = []
#剪枝去除的观点词对应的特征词
for key in list4:
    for fea in set(opinion_to_fea_dict[key]):
        #如果某个特征词对应的观点词只有一个并且就是要删除的观点词，那么去掉这个特征词。
        if len(fea_to_opinion_dict[fea]) == 1:
            del_list.append(fea)
#print(del_list)
        #如果某个特征词对应的不是只有一个观点词，那么查看这个特征词所有的观点词
        else:
            i = 0
            for w in set(fea_to_opinion_dict[fea]):
                print('当前处理的特征是', fea) 
                if w in list4:
                    i += 1
                print('特征中', i)
                #如果要删除的观点词占80%则删除这个特征词。
                if i >= len(set(fea_to_opinion_dict[fea])) * 0.8:
                    print('两个以上',len(set(fea_to_opinion_dict[fea])))
                    del_list.append(fea)    
print(del_list)
```

## 剩下的特征词

```python
key_set = fea_to_opinion_dict.keys() - set(del_list)
```

```python
new_key_list = []
for key in fea_to_opinion_dict:
    if key not in del_list:
        new_key_list.append(key)
print(new_key_list)
new_dict = defaultdict(list)
for w in new_key_list:
    for n in fea_to_opinion_dict[w]:
        if n not in list4:
            new_dict[w].append(n)
print(new_dict.keys())
```

```python
with open('F:/data/new_fea_to_opinion_dict.txt', 'w') as f:
    for key, value in new_dict.items():
        f.write(key + ':[')
        f.write(','.join(value))
        f.write(']' + '\n')
```

```python
with open('F:/data/new_fea.txt', 'w') as f:
    for ele in new_key_list:
        f.write(str(ele))
        f.write('\n')
```

```python
new_opi_to_fea_dict = defaultdict(list)
for w in set(list3) - set(list4):
    new_opi_to_fea_dict[w] = opinion_to_fea_dict[w]
print(new_opi_to_fea_dict)
with open('F:/data/new_opinion_to_fea_dict.txt', 'w') as f:
    for key, value in new_opi_to_fea_dict.items():
        f.write(key + ':[')
        f.write(','.join(value))
        f.write(']' + '\n')
with open('F:/data/new_opinion_to_fea_dict.pkl', 'wb+') as f:
    pickle.dump(new_opi_to_fea_dict, f)
with open('F:/data/new_fea_to_opinion_dict.pkl', 'wb+') as f:
    pickle.dump(new_dict, f)
```
